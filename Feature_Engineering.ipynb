```python
import pandas as pd
import numpy as np
import math
import json
import os

# read in the json files data/portfolio.json
data_dir = '/home/ec2-user/SageMaker/ML_project_starbucksOffer/Data'
portfolio = pd.read_json(os.path.join(data_dir,'portfolio.json'), orient='records', lines=True)
profile = pd.read_json(os.path.join(data_dir,'profile.json'), orient='records', lines=True)
transcript = pd.read_json(os.path.join(data_dir,'transcript.json'), orient='records', lines=True)
```

```python
# Transform portfolio data
portfolio = portfolio.join(pd.get_dummies(portfolio.pop('offer_type')))
channels = pd.DataFrame(portfolio.pop('channels'))\
             .explode('channels').assign(value=lambda x:1)\
             .pivot(columns='channels',values='value').fillna(0)
portfolio = portfolio.join(channels)
portfolio.drop(columns='email')   
#portfolio.index = portfolio.pop('id')
portfolio = portfolio.rename(columns={"id":"offer_id"})
```

```python
portfolio
```

```python
# Transform profile data
missing_gender = profile.gender.isna()
profile = profile.assign(missing_data=missing_gender.astype(int))
profile.gender.mask(missing_gender,'N', inplace=True)
profile = profile.join(pd.get_dummies(profile.pop('gender')))
#profile.index = profile.pop('id')
profile = profile.fillna(0)
profile = profile.rename(columns={"id":"person"})
```

```python
profile
```

```python
transcript = transcript.join(
    pd.DataFrame.from_records(transcript.pop('value')))
transcript.offer_id.update(transcript.pop('offer id'))
transcript
```

```python
# Get offer data set by joining three tables
transcript = transcript.fillna(0)
transcript = transcript[["person","event","time","offer_id","reward"]]
offer_df = pd.merge(transcript.query("event in ('offer received')"),
                    transcript.query("event in ('offer completed')"),
                    on=['offer_id','person'],
                    how='left')
```

```python
hasoffer = offer_df.event_y.notna()

offer_df = offer_df.assign(isAccept=hasoffer.astype(int))[["person","time_x","offer_id","isAccept"]]
```

```python
pd.set_option('display.max_columns', None)

offer_profile_df = pd.merge(
    profile,
    offer_df,
    on ='person', 
    how = 'left'
)
offer_profile_portfolio_df = pd.merge(
    offer_profile_df,
    portfolio,
    on = 'offer_id',
    how = 'left'
)

offer_profile_portfolio_df.pop("missing_data")
offer_profile_portfolio_df.pop("person")
offer_profile_portfolio_df.pop("offer_id")
offer_profile_portfolio_df["became_member_on"] = offer_profile_portfolio_df["became_member_on"].astype(str).str[0:4]
offer_profile_portfolio_df = offer_profile_portfolio_df.fillna(0.0)
outcomes = pd.DataFrame(offer_profile_portfolio_df.pop("isAccept"))
features = offer_profile_portfolio_df
```

```python
from sklearn.model_selection import train_test_split
# Split train test data set
X_train, X_test, y_train, y_test = train_test_split(features, outcomes, test_size=0.2, random_state=42)
```
